```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
train_data = pd.read_csv("../data/train.csv")
test_data = pd.read_csv("../data/test.csv")
```

```python
# 데이터 샘플링 (예: 10% 데이터만 사용)
sample_fraction = 0.01  # 샘플 비율 (1%)
train_data = train_data.sample(frac=sample_fraction, random_state=42)
test_data = test_data.sample(frac=sample_fraction, random_state=42)
```

```python
sns.pairplot(train_data, vars=['Age', 'Vintage'], kind='scatter', diag_kind='kde')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Age 분포 확인
plt.figure(figsize=(10, 6))
sns.histplot(data=train_data, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Age와 Response 간의 관계 분석
plt.figure(figsize=(10, 6))
sns.boxplot(x='Response', y='Age', data=train_data)
plt.title('Age vs Response')
plt.show()
```
